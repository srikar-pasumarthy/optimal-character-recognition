```python
# Create an empty dictionary
label_map = {}

# Iterate over the data and labels
for i in range(len(images)):
    # Get the current label and data point
    label = labels[i]
    point = images[i]

    # If the label doesn't exist in the dictionary, add it
    if label not in label_map:
        label_map[label] = []

    # Add the data point to the list of points associated with the label
    label_map[label].append(point)

# Print the dictionary
print(label_map)
```

```python
import re
def tokens(text):
    "List all the word tokens (consecutive letters) in a text. Normalize to lowercase."
    return re.findall('[a-z]+', text.lower()) 
```
